## Impor library

```python
# RUN-NOW: Jalankan sel ini apa adanya.
import pandas as pd
import numpy as np
from pathlib import Path

print("Library siap")
```

## Tentukan jalur file

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Gunakan Path untuk portabilitas lintas platform
candidate_paths = [
    Path("data/synthetic/synthetic_student_performance.csv"),
    Path("../data/synthetic/synthetic_student_performance.csv"),
    Path("/workspaces/metpen-ai-lab/data/synthetic/synthetic_student_performance.csv"),
]

data_file = next((p for p in candidate_paths if p.exists()), None)

if data_file is None:
    raise FileNotFoundError(
        "File data tidak ditemukan. Periksa folder data/synthetic atau jalankan dari root project."
    )

print(f"✓ File ditemukan: {data_file.resolve()}")
```

## Memuat data ke DataFrame

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Baca CSV menjadi Pandas DataFrame
df = pd.read_csv(data_file)

print("✓ Data berhasil dimuat")
print(f"Bentuk dataset: {df.shape[0]} baris × {df.shape[1]} kolom")
```

## Informasi struktur data

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Tampilkan informasi detail: nama kolom, tipe data, non-null count
df.info()
```

### Apa artinya output di atas?

- **Column**: Nama variabel
- **Non-Null Count**: Berapa banyak nilai yang ada (bukan NaN/hilang)
- **Dtype**: Tipe data (float64 = angka desimal, int64 = integer, object = teks/mixed)

Perhatikan: ada kolom dengan nilai hilang jika `Non-Null Count < total rows`.

## Ringkasan statistik deskriptif dasar

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# describe() menampilkan: count, mean, std, min, 25%, 50%, 75%, max
df.describe()
```

### Interpretasi output di atas

- **count**: Jumlah nilai non-missing untuk setiap kolom
- **mean**: Rata-rata
- **std**: Standar deviasi (ukuran penyebaran)
- **min/max**: Nilai terendah dan tertinggi
- **25%/50%/75%**: Kuartil (persentil ke-25, ke-50, ke-75)

**Apa yang kami perhatikan?** Apakah ada nilai ekstrim atau tidak realistis?

## Periksa tipe data setiap kolom

```python
# RUN-NOW: Jalankan sel ini apa adanya.
print("Tipe data setiap kolom:")
print(df.dtypes)

print("\n" + "=" * 50)
print("Kolom numerik:", df.select_dtypes(include=["number"]).columns.tolist())
print("Kolom kategoris:", df.select_dtypes(include=["object", "string"]).columns.tolist())
```

## Nilai unik dalam kolom kategoris

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Lihat kategori dalam beberapa kolom
print("Nilai unik dalam 'program':")
print(df["program"].unique())
print(f"Jumlah: {df['program'].nunique()}")

print("\nNilai unik dalam 'internet_access':")
print(df["internet_access"].unique())
print(f"Jumlah: {df['internet_access'].nunique()}")

print("\nNilai unik dalam 'ai_tool_usage_frequency':")
print(df["ai_tool_usage_frequency"].unique())
print(f"Jumlah: {df['ai_tool_usage_frequency'].nunique()}")
```

## Hitung frekuensi kategori

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Berapa banyak mahasiswa per program?
print("Distribusi per program:")
print(df["program"].value_counts())

print("\nDistribusi internet_access:")
print(df["internet_access"].value_counts())

print("\nDistribusi passed (lulus/tidak):")
print(df["passed"].value_counts())
```

## Periksa nilai hilang (missing values)

```python
# RUN-NOW: Jalankan sel ini apa adanya.
# Hitung missing values
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    "Kolom": missing.index,
    "Jumlah NaN": missing.values,
    "Persentase": missing_percent.values,
})

# Hanya tampilkan kolom dengan nilai hilang
missing_df = missing_df[missing_df["Jumlah NaN"] > 0].sort_values("Jumlah NaN", ascending=False)

if len(missing_df) == 0:
    print("✓ Tidak ada nilai hilang (missing values) dalam dataset")
else:
    print("Kolom dengan nilai hilang:")
    print(missing_df)
```

## Ringkasan hasil inspeksi

```python
# RUN-NOW: Jalankan sel ini apa adanya.
print("📊 RINGKASAN DATASET")
print(
    f"""\n✓ Ukuran: {df.shape[0]} baris × {df.shape[1]} kolom
✓ Variabel numerik: {len(df.select_dtypes(include=['number']).columns)}
✓ Variabel kategoris: {len(df.select_dtypes(include=['object', 'string']).columns)}
✓ Nilai hilang: {df.isnull().sum().sum()} (dalam {(df.isnull().sum() > 0).sum()} kolom)
✓ Duplikat baris: {df.duplicated().sum()}
"""
)
```
